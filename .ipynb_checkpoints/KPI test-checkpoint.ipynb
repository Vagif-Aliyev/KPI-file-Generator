```python
# import necessary libraries

import pandas as pd
import numpy as np
import os
import datetime as dt
```

```python
# Read the folders to observe inside
files=os.listdir(r"/Users/vagifaliyev/Desktop/Qalmaz_608_DDR")

# Check first 10 file in folder
for i, file in enumerate(files[:10]):
    print(i,".", file)
```

```python
# Create dataframe for operation codes which is given in every 2nd sheet of excel file.
df_ops_code=pd.read_excel(r"/Users/vagifaliyev/Desktop/Qalmaz_608_DDR/"+files[0], sheet_name="Əməliyyatların Kodu")

# Drop undesired columns
df_ops_code.drop(columns=["Unnamed: 0","Əməliyyat və ya gecikmənin açıqlaması","Vaxt"], inplace=True)

# Rename columns
df_ops_code.rename(columns={
    "Əməliyyat və ya gecikmənin adı":"Description",
    "Kod":"Operation Code",
    "Vaxt.1":"Status"
}, inplace=True)
```

```python
# Convert status column's catigrical values from MV to PT (Productive Time) and from QMV to NPT(Non Productive Time)
df_ops_code['Status'].replace({"MV":"PT", "QMV":"NPT"}, inplace=True)

# Confirm the change
df_ops_code.head()
```

```python
#Create empty lists for later use:
ddr_list=[]
time_list=[]
all_ops=[]

# Set a loop for combining individual excel files to single KPI excel file.
for file in files:
    df=pd.read_excel(r"/Users/vagifaliyev/Desktop/Qalmaz_608_DDR/"+file,sheet_name="GM-DDR",skiprows=47)
    
# Rename desired columns
    df.rename(columns={
        "Unnamed: 2": "End",
        "Vaxt": "Start",
        "Əməliyyat kodu":"Operation Code",
        "Unnamed: 7": "Status",
        "Saat": "Time",
        "Gündəlik məlumat":"Daily Operations"
    }, inplace=True)

# Data Cleaning: Drop NaN values and empty columns
    df.drop(columns=['Unnamed: 0','Unnamed: 4','Unnamed: 6','Unnamed: 9', 'Unnamed: 10',
       'Unnamed: 11', 'Unnamed: 12', 'Unnamed: 13', 'Unnamed: 14',
       'Unnamed: 15', 'Unnamed: 16', 'Unnamed: 17', 'Unnamed: 18',
       'Unnamed: 19', 'Unnamed: 20', 'Unnamed: 21', 'Unnamed: 22',
       'Unnamed: 23', 'Unnamed: 24', 'Unnamed: 25', 'Unnamed: 26',
       'Unnamed: 27', 'Unnamed: 28', 'Unnamed: 29', 'Unnamed: 30',
       'Unnamed: 31', 'Unnamed: 32', 'Unnamed: 33'], inplace=True)
    df.dropna(axis=0, how='any', inplace=True)

#  Changing dtype of start column to datetime
    df['Start']=pd.to_datetime(df['Start'])

# Add created dataframes to ddr_list    
    ddr_list.append(df) 
    
# Create daily activity column which joins operation in every excel file
    daily_ops=[]
    joined_description=" "
    for description in df["Daily Operations"]:
        joined_description+=description
    daily_ops.append(joined_description)
    df_daily_ops=pd.DataFrame(daily_ops)
    all_ops.append(df_daily_ops)
    
# Create dataframe of operation times for every operation code for each day
    daily_grouped_hours=df.groupby(by="Operation Code")["Time"].sum()
    df_grouped_hours=pd.DataFrame(daily_grouped_hours).T
    time_list.append(df_grouped_hours)
    
# Create Dataframe of operations
df_all_ops=pd.concat(all_ops, axis=0,ignore_index=True)

# Create Dataframe of time for each operation code for each day
df_all_time=pd.concat(time_list, axis=0,ignore_index=True).fillna(0)

# Combine time and operation dataframes and export final KPI file.
df_final=pd.concat([df_all_ops, df_all_time], axis=1)
df_final.to_excel(r"/Users/vagifaliyev/Desktop/KPI_Demo_v3.xlsx", sheet_name="Operations")
```
